```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
print(os.listdir())
```

```python
df=pd.read_csv('Cleaned_DS_Jobs.csv')
df
```

# Using Salary Estimate  split the maximum and minimum salary.

```python
df[['Min Salary', 'Max Salary']] = df['Salary Estimate'].str.split('-', expand=True).astype(int)# Split into two columns at hyphen
df
```

# the 'Headquarters' and ''Location' split into city and state column

```python
# Extract state (2-letter code) using regex
df['HQ State'] = df['Headquarters'].str.extract(r',\s*([A-Z]{2})')[0]
df['Loc State'] = df['Location'].str.extract(r',\s*([A-Z]{2})')[0]

# Extract city (text before comma)
df['HQ City'] = df['Headquarters'].str.extract(r'^([^,]+)')[0]
df['Loc City'] = df['Location'].str.extract(r'^([^,]+)')[0]
```

```python
df
```

# replace 'Unknown / Non-Applicable' as NaN

```python
df.replace(['Unknown / Non-Applicable', 'Unknown', 'Na'], np.nan, inplace=True)#replace the "Unknown / Non-Applicable"to nan to make more reailble
```

# Ploting  bar chart for Number of job offers per Seniority using columns no_of_job_offers and job_seniority.

```python
# Extract the required columns from the DataFrame
job_simp=df['job_simp']# Number of job offers
seniority=df['seniority']# Seniority categories 
# Create figure with appropriate size
fig=plt.figure(figsize=(12,10))# Slightly more balanced size than 12,10
# Create bar plot
plt.bar(job_simp,seniority)
plt.xlabel('job_simp')
plt.ylabel('seniority')
plt.title('Number of job offers per Seniority')
plt.show()
```

# Ploting bar chart for Number of job offers per Seniority and Sector using columns no_of_job_offers and job_seniority, Sector.

```python
# Group and aggregate the data by seniority and sector
grouped_data=df.groupby(['seniority','Sector'])['job_simp'].sum().reset_index()
# Set up the plot with a more reasonable size 
plt.figure(figsize=(80,50))
# Get unique categories for positioning
for sector in grouped_data['Sector'].unique():
    sector_data=grouped_data[grouped_data['Sector']==sector]
    x=range(len(sector_data))
    plt.bar(x,sector_data['job_simp'],label=sector)
#create bar graph
plt.xticks(range(len(grouped_data['seniority'].unique())),grouped_data['seniority'].unique())
plt.xlabel('Seniority')
plt.ylabel('Number of job offers')
plt.title('Number of job offers per seniority and sector')
plt.legend()
plt.show()
```

# Plot bar chart for Average estimate salary per Sector using column Average estimate salary, sector.

```python
# Clean salary data: remove $ and commas, then split into min/max
df['Salary Estimate']=df['Salary Estimate'].str.replace('[$,]','',regex=True)
df[['Salary Min','Salary Max']]=df['Salary Estimate'].str.split('-',expand=True).astype(float)
# Calculate average salary 
df['Average Salary']=(df['Salary Min']+ df['Salary Max'])/2
# Group by sector and calculate mean average salary
grouped_data=df.groupby('Sector')['Average Salary'].mean().reset_index()
# Create the plot
plt.figure(figsize=(10,5))
plt.bar(grouped_data['Sector'],grouped_data['Average Salary'])
plt.xlabel('Average Estimated salary')
plt.ylabel('Sector')
plt.title('Average estimate salary per Sector')
# Rotate x-axis labels for readability
plt.xticks(rotation=90)
plt.show()
```

# Ploting Scatter plot for Correlation between the average salary and the company rating use columns Average estimate salary, Rating, Seniority.

```python
# Extract the two variables we want to analyze
salary=df['avg_salary']
rating=df['Rating']
plt.figure(figsize=(10,6))# Set figure dimensions
#creat scatter plot 
plt.scatter(rating,salary)
plt.xlabel('Company Rating')
plt.ylabel('Average Estimated Salary')
plt.title('Correlation between the average salary and the company rating')
plt.show()
```

# Plot histplot Distribution of the average estimate salary per type of ownership using Avg_salary_estimate, Type of ownership

```python
#extracting the data
avg_salary=df['avg_salary']
ownership=df['Type of ownership']
# PLOT SETUP
plt.figure(figsize=(20,15))
#create histo plot
sns.histplot(data=df,x=avg_salary,hue=ownership,kde=True)
plt.xlabel('Average Estimated Salary')#x-axis labeling 
plt.ylabel('Count')#y-axis labeling
plt.title('Distribution of the average estimate salary per type of ownership')#title of the graph 
plt.legend(title='Type of Ownership')
plt.show()
```

# Plot boxplot for Distribution of the average salary in each company size using Size, Avg_salary_estimate,

```python
# DATA PREPARATION
avg_salary=df['avg_salary']
company_size=df['Size']
#create boxplot 
sns.boxplot(x=company_size,y=avg_salary)
plt.xlabel('Company Size')#x-axis labeling 
plt.ylabel('Average Estimated Salary')#y-axis labeling
plt.title('Distribution of the average salary in each company size')#title of the graph
plt.xticks(rotation=45)
plt.show()
```

# Ploting  countplot for Number of job offers per Company size using Size, job

```python
# Extract the relevant columns from the dataframe
company_size=df['Size']
job=df['Job Title']# Categorical variable for job titles
plt.figure(figsize=(10,6))#size of plot
#creating count polt 
sns.countplot(x=company_size,hue=job)
plt.xlabel('Company Size')#x-axis labeling
plt.ylabel('Number of job offers')#y-axis labeling 
plt.title('Number of job offers per Company size')#tile of the coutplot
plt.legend(title='Number of job offers per Company size')
plt.xticks(rotation=45)
plt.show()
```
